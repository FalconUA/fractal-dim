```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import cv2
import numpy
import math


%pylab inline
```

```python
def rgb2gray(rgb):        
    return numpy.array(numpy.dot(rgb[...,:3], [0.333, 0.334, 0.333]))
    #return np.array(np.dot(rgb[...,:3], [0.299, 0.587, 0.114]))

initial_image = mpimg.imread("Tests/1_1.bmp")
#initial_image = mpimg.imread("ripple.png")

grayscale_image = rgb2gray(initial_image)
print(grayscale_image)
plt.imshow(grayscale_image, cmap = plt.get_cmap('gray'))
```

```python
def coefficients(x):
    def gen(k):
        #assert k % 2 == 0
        if k % 2 == 1:
            k = k-1
        return (x for x in range(-k//2, k//2) if not x == 0)
    
    def delta_b(x):
        n, m = x.shape        
        assert x.dtype == numpy.complex_        
        a, b, c, d = 0.0, float(n) * float(m), 0.0, 0.0
        for i in gen(n):
            for j in gen(m):
                a += math.log(numpy.abs(x[i][j])) 
                c += math.log(numpy.abs(x[i][j])) * math.log(i**2 + j**2)
                d += math.log(i**2 + j**2) 
                                
        return -1.0 * (a*d - c*b)
    
    def delta_d(x):
        n, m = x.shape        
        assert x.dtype == numpy.complex_        
        a, b, c, d = 0.0, 0.0, 0.0, 0.0
        for i in gen(n):
            for j in gen(m):
                a += math.log(i**2 + j**2)
                b += math.log(numpy.abs(x[i][j])) 
                c += (math.log(i**2 + j**2))**2
                d += math.log(numpy.abs(x[i][j])) * math.log(i**2 + j**2)
                                
        return -0.5 * (a*d - c*b)
    
    def delta(x):
        n, m = x.shape        
        assert x.dtype == numpy.complex_        
        a, b, c, d = 0.0, float(n) * float(m), 0.0, 0.0
        for i in gen(n):
            for j in gen(m):
                a += math.log(i**2 + j**2)
                c += (math.log(i**2 + j**2))**2
                d += math.log(i**2 + j**2)                        
                
        return 0.5 * (a*d - c*b)
    
    b = delta_b(x) / delta(x)
    C = math.exp(delta_d(x) / delta(x))
    return b, C
```

```python
b, c = coefficients(numpy.fft.fft2(grayscale_image))
dim  = 1.0 + (3.0-b)/2.0
print('beta: %f' % b)
print('dim: %f ' % dim)
```

```python
pylab.rcParams['figure.figsize'] = (10, 6)

img = cv2.imread('1_1.jpg',0)
img = cv2.medianBlur(img,5)

ret2,th0 = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
           cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [th0, th1, th2, th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# extrapolation

def extrapolation(img, cell_mask, background_value=0):
    def do_iteration(img, cell_mask, background_value, b_prev, eps=0.0001):
        assert cell_mask.shape == img.shape
        
        #2. perform the fourier transform on the complete image img
        xf = numpy.fft.fft2(img)
                
        #3. estimate coefficients beta(m) and C(m)
        b, c = coefficients(xf)                
        
        #4. if abs(b-b_prev) is less than eps then we say that method is converged
        if not b_prev == None:
            if abs(b - b_prev) < eps:
                return img, b
                
        #5. determine composed spectrum            
        n, m = img.shape                    
        for i in range(-n//2, n//2):
            for j in range(-m//2, m//2):
                a = xf[i][j]
                ang = numpy.angle(a)
                mag_prev = numpy.abs(a)
                mag = c * math.pow((i**2 + j**2), -b/2)
                xf[i][j] = mag * exp(ang*1j)                                        
        
        #6. perform inverse fft (not normalized)
        xi = numpy.fft.ifft2(xf)
        
        #print ('coefficient:',xi.shape[0] * xi.shape[1])
        for i in range(-n//2, n//2):
            for j in range(-m//2, m//2):
                xi[i][j] *= xi.shape[0] * xi.shape[1]
                
        #7. apply extension            
        new_img = numpy.ndarray(img.shape)        
        for i in range(n):
            for j in range(m):
                if cell_mask[i][j] == background_value:
                    new_img[i][j] = xi[i][j].real                    
                else:
                    new_img[i][j] = img[i][j]          
        
        return new_img, b
                
        
    new_img, b_prev = do_iteration(img, cell_mask, background_value, None)
    print('step #0:', b_prev)
    counter = 0    
    while True:
        counter += 1
        new_img, b = do_iteration(new_img, cell_mask, background_value, b_prev)
        print('step #%d:' % counter, b)
        if abs(b-b_prev) < 0.0001:
            break
        b_prev = b             
    
    print('new img:', new_img)
    return new_img, b
```

```python
print("img: ", grayscale_image[80][80])
new_grayscale, b = extrapolation(grayscale_image, th0, 0)
plt.imshow(new_grayscale, cmap = plt.get_cmap('gray')) 

numpy.savetxt('test.txt', new_grayscale)
```
