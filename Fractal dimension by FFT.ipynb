```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
from mpl_toolkits.mplot3d import Axes3D
import cv2
import numpy
import math


%pylab inline
```

```python
img = mpimg.imread("Tests/1_1.bmp")

def rgb2gray(rgb):        
    #return numpy.array(numpy.dot(rgb[...,:3], [0.333, 0.334, 0.333]))
    return np.array(np.dot(rgb[...,:3], [0.299, 0.587, 0.114]))

grayscale = rgb2gray(img)
plt.imshow(grayscale, cmap = plt.get_cmap('gray'))
```

```python
imfft = numpy.fft.fft2(grayscale)

n, m = imfft.shape

n, m = imfft.shape
max_magnitude = 0
for i in range(n):
    for j in range(m): 
        if abs(imfft[i][j]) > max_magnitude and not (i == 0 and j == 0):
            max_magnitude = abs(imfft[i][j])

imfft_gray = numpy.ndarray(imfft.shape, dtype=int_)
for i in range(n):
    for j in range(m):             
        imfft_gray[i][j] = int(255.0 * (abs(imfft[i][j]) / max_magnitude))
        if i == 0 and j == 0:
            imfft_gray[i][j] = 125

plt.imshow(imfft_gray, cmap = plt.get_cmap('gray'))
```

```python
im = numpy.fft.ifft2(imfft)

n, m = imfft.shape
gray = numpy.ndarray(imfft.shape)
for i in range(n):
    for j in range(m):         
        gray[i][j] = im[i][j].real
            
plt.imshow(gray, cmap = plt.get_cmap('gray'))
```
