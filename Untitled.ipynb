```python
im0 = numpy.ndarray(shape=(50,50))
for i in range(im0.shape[0]):
    for j in range(im0.shape[1]):
        im0[i][j] = random.randint(0, 256)
plt.imshow(im0, cmap = plt.get_cmap('gray'))
```

```python
im0fft = numpy.fft.fft2(im0)

im0fft_gray = numpy.ndarray(shape=im0fft.shape, dtype=float_)
for i in range(-im0.shape[0]//2, im0.shape[0]//2):
    for j in range(-im0.shape[1]//2, im0.shape[1]//2):         
        im0fft_gray[i][j] = numpy.abs(im0fft[i][j])
        if i == 0 and j == 0:
            im0fft_gray[i][j] = 0

plt.imshow(im0fft_gray, cmap = plt.get_cmap('gray'))
```
